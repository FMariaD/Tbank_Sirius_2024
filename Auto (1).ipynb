```python
from diffusers import DiffusionPipeline
from PIL import Image, ImageOps
import requests
from io import BytesIO
import torch
from transparent_background import Remover
import os
import tqdm
import numpy as np
import matplotlib.pyplot as plt
```

```python
model_id = "yahoo-inc/photo-background-generation"
pipeline = DiffusionPipeline.from_pretrained(model_id, custom_pipeline=model_id)
#pipeline = pipeline.to('cuda')
```

```python
def resize_with_padding(img, expected_size):
    img.thumbnail((expected_size[0], expected_size[1]))
    # print(img.size)
    delta_width = expected_size[0] - img.size[0]
    delta_height = expected_size[1] - img.size[1]
    pad_width = delta_width // 2
    pad_height = delta_height // 2
    padding = (pad_width, pad_height, delta_width - pad_width, delta_height - pad_height)
    return ImageOps.expand(img, padding), padding
```

```python
# Load background detection model
remover = Remover() # default setting
remover = Remover(mode='base') # nightly release checkpoint
```

```python
def replace_background(img_path, prompt):
    img = Image.open(img_path)
    img, padding = resize_with_padding(img, (512, 512))

    # Get foreground mask
    fg_mask = remover.process(img, type='map') # default setting - transparent background

    seed = 13
    mask = ImageOps.invert(fg_mask)
    generator = torch.Generator(device='cpu').manual_seed(seed)
    cond_scale = 1.0
    with torch.autocast("cpu"):
        controlnet_image = pipeline(prompt=prompt,
                                    image=img,
                                    mask_image=mask,
                                    control_image=mask,
                                    num_images_per_prompt=1,
                                    generator=generator,
                                    num_inference_steps=30,
                                    guess_mode=False,
                                    controlnet_conditioning_scale=cond_scale)
    return controlnet_image.images[0], padding
```

```python
BATCH_SIZE = 30
DIR_PATH = r"/content/drive/MyDrive/images"
DIR2SAVE = r"/content/drive/MyDrive/images_made"

img_paths = [os.path.join(DIR_PATH, p) for p in os.listdir(DIR_PATH)]
```

```python
# TryOut

preds = {}
PROMPT = 'empty plain one-colour %s wall'
color = 'white'

for i, img_path in enumerate(img_paths):
    idx = img_path.split('/')[-1].split('.')[0]

    if idx+'.jpg' not in os.listdir(DIR2SAVE):

        if int(idx) % 2 == 0:
            color = 'grey'
        else:
            color = 'white'
        pred, pads = replace_background(img_path, PROMPT % color)

        depadding = (pads[0], pads[1], pred.size[0]-pads[2], pred.size[1]-pads[3])
        pred.crop(depadding).save(os.path.join(DIR2SAVE, idx+'.jpg'))
```
