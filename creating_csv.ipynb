```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import os
```

```python
IMG_DIR = r'C:\Users\FMari\Sirius2024\images'
FIN_DIR = r'C:\Users\FMari\Sirius2024\images_final'
SH_DIR = r'C:\Users\FMari\Sirius2024\short_captions'
LG_DIR = r'C:\Users\FMari\Sirius2024\long_captions'

DIR2SAVE = r'C:\Users\FMari\Sirius2024\ANSWER'
```

```python
for path1 in os.listdir(IMG_DIR):
    try:
        plt.figure()
        name = path1.split('.')[0]
        
        #subplot(r,c) provide the no. of rows and columns
        f, axarr = plt.subplots(2,1) 
        
        # use the created array to output your multiple images. In this case I have stacked 4 images vertically
        axarr[0].imshow(Image.open(os.path.join(IMG_DIR, path1)))
        axarr[1].imshow(Image.open(os.path.join(FIN_DIR, path1)))

        with open(os.path.join(SH_DIR, name+'.txt'), 'r') as file:
            short = file.read()
        with open(os.path.join(LG_DIR, name+'.txt'), 'r') as file:
            long = file.read()
        
        f.text(1, 1, short+'\n'+long)
        
        f.show()
    except:
        pass

        
    break
```
